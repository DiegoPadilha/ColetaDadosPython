```python
mydb = mysql.connector.connect(
    host='localhost',
    user='root',
    password='igti',
    database='bootcamp'
)

print(mydb)

mycursor = mydb.cursor()
```

```python
import csv

#Leitura de arquivo sem utilizar biblioteca Pandas (CSV)

filename = 'C:\Bootcamp\Datasets\CSV\caracteristicaImovel.csv'

with open(filename, 'r') as incsvfile:
    reader = csv.reader(incsvfile, delimiter=',')
    next(reader, None)  # skip the headers
    for line in reader:
        #print(line)
        #print(line[1])
        mycursor.execute ("INSERT INTO caracteristicageralimovel (idcaracteristicasGerais,idImovel,temCaracteristica)\
                  VALUES (%s, %s, %s)",line)        
mydb.commit()

print(mycursor.rowcount, "registro(s) inserido(s).")

#db.close()
incsvfile.close()
```

```python
#Leitura de arquivo sem utilizar biblioteca Pandas (TXT)

filename = 'C:\Bootcamp\Datasets\TXT\caracteristicaImovel.txt'

with open(filename, 'r') as intxtfile:
    for line in intxtfile:
      a, b, c  = line.split(",")
      if a == "idcaracteristicasGerais":
         continue
    
      print(a, b, c)
      stmt = "INSERT INTO caracteristicageralimovel (idcaracteristicasGerais,idImovel,temCaracteristica)\
         VALUES (%s, %s, %s)" % (a, b, c)
      mycursor.execute (stmt)
        
mydb.commit()

print(mycursor.rowcount, "registro(s) inserido(s).")

#db.close()
incsvfile.close()
```
